## 분석과 시각화를 위한 tidy data 만들기

* 컬럼에 있는 데이터를 행으로 보낸다. 그래야 컬럼을 통해 기간, 분기 라는 컬럼을 만들어서 기간별 데이터를 분석해보고 시각화 할 수 있다.

```python
df_raw.columns
```

```python
df_raw.values
```

```python
# id_vars만 지정해주면 나머지 컬럼 값은 알아서 value_vars로 지정된다.

df_raw.melt(id_vars=['국가(대륙)별', '상품군별', '판매유형별'])
```

```python
df = df_raw.melt(id_vars=['국가(대륙)별', '상품군별', '판매유형별'], var_name='기간', value_name='백만원')
df
```

```python
df.head()
```

```python
df.info()
```

## 데이터 전처리

### 기간에서 연도를 분리하기

```python
df['기간']
```

```python
type("2021 2/4 p)".split()[0])
```

```python
"2021 2/4 p)".split()[0]
```

```python
int("2021 2/4 p)".split()[0])
```

```python
# map은 우리가 그 안에 익명함수 지정해주면 그게 실행됨.
# x는 df['기간']

df['연도']=df['기간'].map(lambda x : int(x.split()[0]))
```

```python
df.head()
```

```python
int("2021 2/4 p)".split()[1].split("/")[0])
```

```python
df['분기']=df['기간'].map(lambda x : int(x.split()[1].split("/")[0]))
```

```python
df.head()
```

```python
# - 때문에 오브젝트 타입으로 나온다. 자료없음.
df['백만원']
```

```python
df['백만원'] = df['백만원'].replace("-", np.nan).astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python
# 합계 데이터 필요없어서 제거

df = df[(df['국가(대륙)별'] != '합계') & (df['상품군별'] != '합계')].copy()
df.info()
```

```python
df.isnull().sum()
```
